```python
df['Memory'] = df['Memory'].astype(str).replace('\.0', '', regex=True)
df["Memory"] = df["Memory"].str.replace('GB', '')
df["Memory"] = df["Memory"].str.replace('TB', '000')
new = df["Memory"].str.split("+", n = 1, expand = True)

df["first"]= new[0]
df["first"]=df["first"].str.strip()

df["second"]= new[1]

df["Layer1HDD"] = df["first"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer1SSD"] = df["first"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer1Hybrid"] = df["first"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer1Flash_Storage"] = df["first"].apply(lambda x: 1 if "Flash Storage" in x else 0)

df['first'] = df['first'].str.replace(r'\D', '', regex=True)

df["second"].fillna("0", inplace = True)

df["Layer2HDD"] = df["second"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer2SSD"] = df["second"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer2Hybrid"] = df["second"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer2Flash_Storage"] = df["second"].apply(lambda x: 1 if "Flash Storage" in x else 0)

df['second'] = df['second'].str.replace(r'\D', '', regex=True)

df["first"] = df["first"].astype(int)
df["second"] = df["second"].astype(int)

df["HDD"]=(df["first"]*df["Layer1HDD"]+df["second"]*df["Layer2HDD"])
df["SSD"]=(df["first"]*df["Layer1SSD"]+df["second"]*df["Layer2SSD"])
df["Hybrid"]=(df["first"]*df["Layer1Hybrid"]+df["second"]*df["Layer2Hybrid"])
df["Flash_Storage"]=(df["first"]*df["Layer1Flash_Storage"]+df["second"]*df["Layer2Flash_Storage"])

df.drop(columns=['first', 'second', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid',
       'Layer1Flash_Storage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid',
       'Layer2Flash_Storage'],inplace=True)
```

### Linear Regression

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error

# Categorical columns by index (as in your code)
cat_cols = [0, 1, 7, 10, 11]

step1 = ColumnTransformer(
    transformers=[
        (
            'col_tnf',
            OneHotEncoder(
                sparse_output=False,   # <- new param (replaces sparse=False)
                drop='first',
                handle_unknown='ignore'
            ),
            cat_cols
        )
    ],
    remainder='passthrough'
)

step2 = LinearRegression()

pipe = Pipeline(steps=[
    ('step1', step1),
    ('step2', step2)
])

pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)

print("R² score:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
```

### Ridge Regression

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score, mean_absolute_error

# ---- Specify your columns ----
# Indices of categorical columns (from your screenshot)
cat_cols = [0, 1, 7, 10, 11]
# All other columns are numeric; if you know their indices explicitly, list them.
# Otherwise, infer them from X_train:
num_cols = [i for i in range(X_train.shape[1]) if i not in cat_cols]

# ---- Preprocess: OHE for categorical, scaling for numeric ----
preprocess = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(
            sparse_output=False,   # <— replaces sparse=False
            drop='first',
            handle_unknown='ignore'
        ), cat_cols),
        ('num', StandardScaler(), num_cols)
    ],
    remainder='drop'  # or 'passthrough' if you want to keep any unlisted columns
)

# ---- Model ----
ridge = Ridge(alpha=10, random_state=42)

# ---- Pipeline ----
pipe = Pipeline(steps=[
    ('prep', preprocess),
    ('model', ridge)
])

# ---- Fit & evaluate ----
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)

print("R² score:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
```

### Lasso Regression

```python
# --- Lasso Regression: robust pipeline ---

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score, mean_absolute_error
import numpy as np
import pandas as pd

# 0) Ensure y is 1-D numeric
# If y_train / y_test are DataFrames with a single column, flatten them:
if isinstance(y_train, (pd.DataFrame, np.ndarray)) and y_train.ndim > 1:
    y_train = np.ravel(y_train)
if isinstance(y_test, (pd.DataFrame, np.ndarray)) and y_test.ndim > 1:
    y_test = np.ravel(y_test)

# 1) Define columns
cat_cols = [0, 1, 7, 10, 11]                         # your categorical column indices
num_cols = [i for i in range(X_train.shape[1]) if i not in cat_cols]

# 2) Optional: quick NaN check (fail early with a clear message)
def assert_no_nan(name, arr):
    if isinstance(arr, pd.DataFrame):
        has_nan = arr.isna().any().any()
    else:
        has_nan = np.isnan(arr).any()
    if has_nan:
        raise ValueError(f"{name} contains NaNs. Impute or drop them before fitting.")

assert_no_nan("X_train", X_train)
assert_no_nan("y_train", y_train)
assert_no_nan("X_test", X_test)

# 3) Preprocess: OHE for categoricals (dense), scaling for numerics
preprocess = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(
            sparse_output=False,   # <- use this on sklearn >= 1.2 (replaces sparse=False)
            drop='first',
            handle_unknown='ignore'
        ), cat_cols),
        ('num', StandardScaler(), num_cols)
    ],
    remainder='drop'  # or 'passthrough' if you intentionally keep other columns
)

# 4) Model
lasso = Lasso(alpha=0.001, max_iter=10000, random_state=42)

# 5) Pipeline
pipe = Pipeline(steps=[
    ('prep', preprocess),
    ('model', lasso)
])

# 6) Fit & evaluate
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)

print('R2 score', r2_score(y_test, y_pred))
print('MAE', mean_absolute_error(y_test, y_pred))
```
